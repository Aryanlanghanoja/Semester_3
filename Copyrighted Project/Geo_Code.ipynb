```python
from geopy.geocoders import Nominatim


def get_coordinates(state):
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.geocode(state)
    if location:
        return (location.latitude, location.longitude)
    else:
        return None


def main():
    state = "Henderson"
    coordinates = get_coordinates(state)
    if coordinates:
        print(
            f"The coordinates of {state} are: Latitude {coordinates[0]}, Longitude {coordinates[1]}"
        )
    else:
        print("Coordinates not found for the given state.")


if __name__ == "__main__":
    main()
```
